# Log Transformation

```python
import skimage.io as io
import matplotlib.pyplot as plt
# Import NumPy library
import numpy as np
```

```python

# Load the image as a NumPy array
image = io.imread("D:/Teaching/2023-2024 First Term/image multi/chapters images/DIP3E_Original_Images_CH03/log.tif")
plt.imshow(image, cmap="gray")
plt.show()
```

```python
# Apply the log transformation
# image = 1e-10 + image
new_image = np.log(image + 1)
plt.imshow(new_image, cmap="gray")
plt.show()
```

# Gamma Transformation

```python
# Load the image as a NumPy array
image = io.imread("D:/Teaching/2023-2024 First Term/image multi/chapters images/DIP3E_Original_Images_CH03/Fig0308(a)(fractured_spine).tif")
plt.imshow(image, cmap="gray")
```

```python
# Define the gamma value
gamma = 0.5
new_image = np.power(image, gamma)
# show new image
plt.subplot(1,3,1)
plt.imshow(new_image, cmap="gray")
plt.title('gamma=0.5')

gamma = 0.4
new_image = np.power(image, gamma)
plt.subplot(1,3,2)
plt.imshow(new_image, cmap="gray")
plt.title('gamma=0.4')

gamma = 0.3
new_image = np.power(image, gamma)
plt.subplot(1,3,3)
plt.imshow(new_image, cmap="gray")
plt.title('gamma=0.3')
```
