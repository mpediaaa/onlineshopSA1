```python
import skimage.io as io
import matplotlib.pyplot as plt
```

```python
f = io.imread("C:/Users/hp/OneDrive/Pictures/Saved Pictures/color.jfif")
plt.imshow(f)
```

```python
plt.imshow(f)
plt.title('festival')
plt.show()
```

```python
print(f.shape)
```

```python
import skimage.color as color
g = color.rgb2gray(f)
plt.imshow(g, cmap="gray")
plt.title('festival')
plt.show()
```

```python
print(f"Min : {g.min()}")
print(f"Max : {g.max()}")
```

```python
# redefined colors
plt.imshow(g, vmin=0.0, vmax=0.5, cmap="gray");
plt.title("Intensity range = [0.0, 0.5]");
plt.show()
```
