```python
import skimage.io as io
import matplotlib.pyplot as plt
import skimage.filters as fltr
import numpy as np
from scipy import ndimage
```

```python
image = io.imread("eye.tif")
# print(image.size)
# Apply the mean filter with a 3x3 kernel
mean_filter = ndimage.uniform_filter(image, size=3)
# Apply the median filter with a 3x3 kernel
median_filter = ndimage.median_filter(image, size=3)
# Apply the min filter with a 3x3 kernel
min_filter = ndimage.minimum_filter(image, size=3)
# Apply the max filter with a 3x3 kernel
max_filter = ndimage.maximum_filter(image, size=3)
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2,3,1)
plt.imshow(image, 'gray')
plt.title('Original image')
plt.subplot(232)
plt.imshow(mean_filter, 'gray')
plt.title('Mean filter')
plt.subplot(233)
plt.imshow(median_filter, 'gray')
plt.title('Median filter')
plt.subplot(234)
plt.imshow(min_filter, 'gray')
plt.title('Min filter')
plt.subplot(235)
plt.imshow(max_filter, 'gray')
plt.title('Max filter')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage.filters

# Reading of the image into numpy array:
A0 = plt.imread('Fig0338(a)(blurry_moon).tif')

# Map values to the (0, 255) range:
A0 = (A0 - np.amin(A0))*255.0 /(np.amax(A0)-np.amin(A0)) 

# Kernel for negative Laplacian:
# kernel = [[-1,-1,-1],[-1,8,-1],[-1,-1,-1]]
kernel      = np.ones((3,3))*(-1)
kernel[1,1] = 8

# Convolution of the image with the Laplacian kernel:
Lap = scipy.ndimage.convolve(A0, kernel)

# Add Laplacian edges to the original image:
A = A0 + Lap 
# Set negative values to 0, values over 255 to 255:
A = np.clip(A, 0, 255)

# Image the result:
plt.figure(figsize=(12, 8))
plt.subplot(131)
plt.imshow(A0, 'gray')
plt.title('Original image')
plt.axis('off')
plt.subplot(132)
plt.imshow(abs(Lap), 'gray')
plt.title('laplace')
plt.axis('off')
plt.subplot(133)
plt.imshow(A, 'gray')
plt.title('Sharpen Image')
plt.axis('off')

plt.show()
```
