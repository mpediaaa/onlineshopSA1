```python
import skimage.io as io
import skimage.exposure as ex
import matplotlib.pyplot as plt
import skimage.color as clr
import numpy as np
```

```python
# Read the image as grayscale
img = io.imread('D:/Teaching/2023-2024 First Term/image multi/chapters images/DIP3E_Original_Images_CH03/low.tif')
```

```python

# Display the original and stretched images
io.imshow(img, cmap="gray")
io.show()
```

```python
plt.hist( np.ravel(img), bins=256)
plt.show()
```

```python
# Find the minimum and maximum pixel values
minval = img.min()
print(f"Min : {minval}")
maxval = img.max()
print(f"Max : {maxval}")
```

```python

# Perform contrast stretching
img_stretched = ex.rescale_intensity(img)
```

```python
plt.imshow(img_stretched, cmap="gray")
plt.show()
```
